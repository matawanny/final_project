```python
import yfinance as yf
import pandas as pd
import datetime
```

```python
tickets=['FB', 'SHOP', 'BYND', 'NFLX', 'PINS', 'SQ', 'TTD', 'OKTA', 'SNAP', 'DDOG']
```

```python
def listToString(s):  
    
    # initialize an empty string 
    str1 = " " 
    
    # return string   
    return (str1.join(s)) 
```

```python
tickets_str = listToString(tickets)
print(tickets_str)
```

```python
data = yf.download(  # or pdr.get_data_yahoo(...
        # tickers list or string as well
        tickers = f"{tickets_str}",

        # use "period" instead of start/end
        # valid periods: 1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max
        # (optional, default is '1mo')
        period = "5d",

        # fetch data by interval (including intraday if period < 60 days)
        # valid intervals: 1m,2m,5m,15m,30m,60m,90m,1h,1d,5d,1wk,1mo,3mo
        # (optional, default is '1d')
        interval = "1m",

        # group by ticker (to access via data['SPY'])
        # (optional, default is 'column')
        group_by = 'ticker',

        # adjust all OHLC automatically
        # (optional, default is False)
        auto_adjust = False,

        # download pre/post regular market hours data
        # (optional, default is False)
        prepost = False,

        # use threads for mass downloading? (True/False/Integer)
        # (optional, default is True)
        threads = True,

        # proxy URL scheme use use when downloading?
        # (optional, default is None)
        proxy = None
    )
```

```python
print(data.head(5))
```

```python
data=data.loc['2020-05-14':'2020-05-14']
```

```python
df = pd.DataFrame(columns=['high', 'low', 'ts', 'name'])
```

```python
for ticket in tickets:
    for index, row in data.iterrows():
        new_row={'high':row[ticket]['High'], 'low':row[ticket]['Low'], 'ts':index, 'name':f'{ticket}'}
        df=df.append(new_row, ignore_index=True)
```

```python
df.to_json(orient='records')
```
